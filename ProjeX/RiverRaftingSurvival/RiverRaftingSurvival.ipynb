```python
#load dataset
dataset=pd.read_csv('Survival.csv')
#summarise dataset
print(dataset.shape)
print(dataset.head(5))
```

```python
#mapping text data to binary
income_set=set(dataset['Sex'])
dataset['Sex']=dataset['Sex'].map({'female':0,'male':1}).astype(int)
print(dataset.head(5))
```

```python
#segregation of data
#independent 
X = dataset.drop('Survived',axis='columns')
#dependent
Y = dataset.Survived
```

```python
#removing the null values
X.columns[X.isna().any()]
```

```python
X.Age = X.Age.fillna(X.Age.mean())
```

```python
#recheck for null values
X.columns[X.isna().any()]
```

```python
#train n test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size = 0.25, random_state = 0)
```

```python
#training
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)
```

```python
#predictions
pclassNo = int(input("Enter person's pclassNo: "))
gender = int(input("Enter person's gender (0:F,1:M): "))
age = int(input("Enter person's age: "))
fare = float(input("Enter person's fare: "))
person = [[pclassNo,gender,age,fare]]
result = model.predict(person)

if result == 1:
  print("Person might survive the adventure.")
else:
  print("RIP")
```

```python
#testing
y_pred = model.predict(X_test)
print(np.column_stack((y_pred,Y_test)))
```

```python
#accuracy
from sklearn.metrics import accuracy_score
x = accuracy_score(Y_test,y_pred)*100
print("Accuracy of the model is: ")
print(x)
```
